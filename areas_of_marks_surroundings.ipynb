```python
import cv2
import pandas as pd

import os
from glob import glob
from pathlib import Path
```

```python
DATA_DIR = "C:\\Users\\Lenovo\\1navigation\\circles_2024_marks_surr\\cut"
```

```python
PREPROC_DATA_DIR = 'downsamp_preprocessed_dataset'

# 1. open files
dataset_names = sorted(glob(os.path.join(DATA_DIR, '*.png')))
images = {}
for nam in dataset_names:
    print(nam)
    k = Path(nam).stem
    k = k.replace(' = ', '')
    k = k.replace(' ', '_')
    k = k.replace(',', '')
    im = cv2.imread(nam)
    images[k] = im
```

```python
#count areas of different colors
colors_areas = pd.DataFrame()


for n, im in images.items():
    coords = []
    for i in range(len(im)):
        for j in range(len(im[i])):
            coords.append([i,j,im[i][j][2],im[i][j][1],im[i][j][0]])
    df = pd.DataFrame(coords, columns =['y', 'x', 'red', 'green', 'blue'])
    df['color'] = 'Not defined'
    df.loc[(df['green']>200)&(df['blue']<100)&(df['red']<100), 'color'] = 'green' 
    df.loc[(df['blue']>200)&(df['green']<100)&(df['red']<100), 'color'] = 'blue' 
    df.loc[(df['red']>200)&(df['green']>200)&(df['blue']>200), 'color'] = 'white'
    
    lildf = pd.DataFrame({'name_of_picture':[n,n,n,n], 'color':['green','blue','white','not defined'], \
                             'area_perc': [((len(df.loc[df['color']=='green']) / len(df)) * 100), (len(df.loc[df['color']=='blue']) / len(df) * 100),\
                                           ((len(df.loc[df['color']=='white']) / len(df))*100), (len(df.loc[df['color']=='Not defined']) / len(df)* 100)]})
    colors_areas = colors_areas.append(lildf)

print(colors_areas)

```

```python
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print(colors_areas)
```

```python
colors_areas.to_csv('marks_areas_colors_areas.csv')
```
